# Let the Complete Guesswork begin
The previous part was either fully based on data or based on some relatively solid estimations given centuries of observation. This will not be that.
The goal of this section is to model the adoption, development and evolution of an entirely new, many claim revolutionary, technology.
We will try to transfer what we as humans have learnt about the evolution of technology and try to apply it to Generative AI. However, there are many caveats:
1. We can model AI adoption rate relatively easily, we just try to fit the data we have about the number of AI users to development curves of past technologies. In my opinion, it will likely be the case that GenAI will creep into most aspects of human life whether we want it to or not.

```python
import numpy as np
from utils import plot_against_data
def ai_words_en(
    t,                          # year or array of years, e.g. 2025 or np.linspace(2023, 2050, 100)
    W_max = 3.0e12,
    k     = 2.101447,
    t_mid = 2026.5,
    eng_start = 0.80,
    eng_decay = 0.02,
    eng_floor = 0.50,
    txt_shift_start = 2028,
    txt_slope  = 0.01,
    txt_floor  = 0.60,
    bound='baseline'
):
    """
    Estimate of *unique* English AI-generated words per day (visible to humans).
    Works with scalars or NumPy arrays.
    """
    t = np.asarray(t, dtype=float)          # vectorise

    # 1) total visible AI text (all languages)
    all_text = W_max / (1 + np.exp(-k * (t - t_mid)))

    # 2) English share (clamped to floor)
    eng_share = np.maximum(eng_floor, eng_start - eng_decay * (t - 2023))

    # 3) Text share vs voice/other modalities (clamped to floor)
    txt_share = np.maximum(
        txt_floor,
        1 - txt_slope * np.maximum(0, t - txt_shift_start)
    )

    result = all_text * txt_share
    if bound=='upper':
        result *= 2
    elif bound=='lower':
        result /= 2
    return result if result.size > 1 else result.item()   # return scalar when given scalar

years = np.linspace(2000, 2050)

fig = plot_against_data(ai_words_en, years, [],[],"Words Generated by AI","#Words")
fig.show()
```
